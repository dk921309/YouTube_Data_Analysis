# Stage 2:Data cleaning

```python
#To remove some columns which we don't require
df=df.drop(['tags','thumbnail_link','description'],axis=1)
df.head() #It will show top 5 rows of dataset
```

```python
#To see the shape of dataset i.e, rows and columns
df.shape # we found that here rows are 37352
```

```python
duplicate_rows=df[df.duplicated()]  #to calculate no. of duplicate rows in dataframe
print("no. of duplicate rows in dataframe: ",duplicate_rows.shape)
```

```python
# To drop the duplicate rows in the dataset 
df=df.drop_duplicates()
df.shape  # It will show lesser no. of row because duplicates enteries has Deleted
```

```python
#To see how many Null values column wise
print(df.isnull().sum()) # it is showing that comment count has two null values
```

```python
#To drop the rows which have Null values
df=df.dropna()
df.count() # It will shows no. of rows in each column after deleting Null value rows which is 2 less in counting
```

```python
#Now again check for Null values in each column, that must be 0
print(df.isnull().sum())
```

# Stage 3: Exploratory Data Analysis

```python
df['publish_time']= pd.to_datetime(df.publish_time) # It will show publish_time in date time format
df.head()
```

```python
list(df) # it will cosider dataframe as list bcoz further operations are easy on lists
```

```python
df[df.isna().any(axis=1)] # To check either any column still have Null values
```

```python
# To convert views, likes and dislikes in millions
numeric_dtype=['views','likes','dislikes']
for i in numeric_dtype:
    df[i]=df[i]/1000000
df
```

```python
#sorting data by likes in descending order
df.sort_values("likes",axis=0,ascending=False,inplace=True,na_position='last')
f1=df.head(20)
f1
```

```python
fig, ax = plt.subplots()
ax.barh(range(f1.shape[0]),f1['likes'])
ax.set_yticks(range(f1.shape[0]))
ax.set_yticklabels(f1['title'])
ax.invert_yaxis()   #labels are from top  to bottom
ax.set_xlabel('likes in Millions')
plt.show()
```

```python
#sorting data by views in descending order
df.sort_values("views",axis=0,ascending=False,inplace=True,na_position='last')
f2=df.head(20)
f2
```

```python
fig, ax = plt.subplots()     # for subplots of types of horizontal bargraph
ax.barh(range(f2.shape[0]),f2['likes'])
ax.set_yticks(range(f2.shape[0]))   # To set the range
ax.set_yticklabels(f2['title'])  # to set the label as title of the videos
ax.invert_yaxis()   #labels are from top  to bottom
ax.set_xlabel('Views in Millions')
plt.show()
```

```python
# To see the top 10 videos of "Zee Yuva" channel
df_videos_top=df[df['channel_title'].str.contains('Zee Yuva')].sort_values(by='views',ascending=False).head(10) # a datframe is created which has channel name as "Zee Yuva" and it videos views in descending order
ax=df_videos_top.plot(kind='bar',x='title',y='views',title='top 10 videos of Zee Yuva Channnel')
ax.set_xlabel('Video Name')
ax.set_ylabel('Views in Millions')
```

```python
#to find the trending videos 'views' with their 'likes' ratio.
sns.jointplot(x='views',y='likes',data=df,kind='reg') # it will show jointplot between Views and Likes of video in regression form, using seaborn
```
